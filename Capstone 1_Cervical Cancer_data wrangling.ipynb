```python
#importing relevant libraries

import pandas as pd
import numpy as np
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
#importing the file as a dataframe

df = pd.read_csv('kag_risk_factors_cervical_cancer.csv', na_values = ["?"])
```

```python
df.head()
```

```python
df.info()
```

```python
#Finding out the total missing values

missing = pd.concat([df.isnull().sum(), 100*df.isnull().mean()], axis = 1)
missing.columns = ['count', '%']
missing.sort_values(by = ['count','%'], ascending = True)
```

There are 2 columns with more than 90% missing values

```python
df.describe()
```

```python
#Drop the columns that have missing values higher than 90%

col_90 = df.columns[df.isna().sum()/df.shape[0] > 0.9]
df.drop(col_90, axis = 1, inplace = True)
df.isna().sum()/df.shape[0] *100
```

```python
df.shape
```

```python
#removing rows with more than 50% missing values

df = df.loc[df.isna().sum(axis = 1) <=df.shape[1]/2]
df.reset_index(drop = True, inplace = True)
df.isna().sum()
```

```python
df.shape
```

Separating the binary and the non-binary data in the dataframe

```python
#separating the binary and the non-binary data

df_binary = df.loc[:,df.nunique() <=2]
df_nonbinary = df.loc[:,df.nunique() > 2]
df_binary.shape,df_nonbinary.shape
```

Check the missing values in binary

```python
df_binary.isna().sum()
```

```python
# Filling the missing values in the binary data

col_bin = df_binary.columns[df_binary.isna().any()]
for col in col_bin:
    df_binary[col].fillna((df_binary[col].mode()[0]), inplace = True)

df_binary.isna().sum()
```

Check the missing values in non binary

```python
df_nonbinary.isna().sum()
```

```python
#Filling the misssing values in the non binary data

col_nbin = df_nonbinary.columns[df_nonbinary.isna().any()]
for col in col_nbin:
    df_nonbinary[col].fillna((df_nonbinary[col].median()), inplace = True)

df_nonbinary.isna().sum()
```

```python
#Removing columns with the same values in all rows

df_binary = df_binary.loc[:, (df_binary!= df_binary.loc[0]).any()]
df_nonbinary = df_nonbinary.loc[:, (df_nonbinary!= df_nonbinary.loc[0]).any()]
```

```python
#remove duplicate columns

df_binary = df_binary.T.drop_duplicates().T
df_nonbinary = df_nonbinary.T.drop_duplicates().T
```
